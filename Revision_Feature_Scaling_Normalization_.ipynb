```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
total_bill=list(df['total_bill'])
```

```python
import numpy as np
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
mean,std
```

```python
normalized_data=[]
for i in total_bill:
  z_score=(i-mean)/std
  normalized_data.append(z_score)

print(normalized_data)
```

```python
sns.histplot(df['total_bill'],kde=True)
```

```python
sns.histplot(normalized_data,kde=True)
```

```python
## let's apply  standardization in new dataset


import seaborn as sns
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
normalized_iris = scaler.fit_transform(df[["sepal_length","sepal_width","petal_length","petal_width"]])
```

```python
## plot without standardize  data

import matplotlib.pyplot as plt

plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
sns.histplot(df['sepal_length'],kde=True)
plt.subplot(1,3,2)
sns.histplot(df['sepal_width'],kde=True)
plt.subplot(1,3,3)
sns.histplot(df['petal_length'],kde=True)
```

```python
## noe plot standardize data

plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
sns.histplot(normalized_iris[:,0],kde=True,color="red")
plt.subplot(1,3,2)
sns.histplot(normalized_iris[:,1],kde=True,color="green")
plt.subplot(1,3,3)
sns.histplot(normalized_iris[:,2],kde=True,color="blue")
```
