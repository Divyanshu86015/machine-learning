#let's create a categorical feature dataframe


```python
import pandas as pd
```

```python
df=pd.DataFrame({"city":["Sultanpur","Delhi","lucknow","lucknow","Sultanpur","Sultanpur","Delhi","Delhi","lucknow"]})
```

```python
df
```

# convert categorical feature into numerical feature with the help of One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#create instance of OneHotEncoder
encoder=OneHotEncoder()
```

```python
encoder.fit(df[["city"]])
```

```python
encoded_data=encoder.transform(df[["city"]])
```

```python
encoded_data
```

```python
encoded_data.toarray()
```

```python
#conver encoded_data into Datarframe
df1=pd.DataFrame(encoded_data.toarray(),columns=encoder.get_feature_names_out())
```

```python
df1
```

```python
## concat in real data
df=pd.concat([df,df1],axis=1)
```

```python
df
```
