```python
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

#**Now perform unit vetcors normalization with the help of sklearn library**

```python
from sklearn.preprocessing import normalize
```

```python
df.columns
```

```python
normalized_data=normalize(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
normalized_data
```

```python
## convert  into dataframe

import pandas as pd
```

```python
df2= pd.DataFrame(normalized_data,columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```

```python
df2
```

```python
## plot datapoint without normalize (unit vertor )

import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,3))
plt.subplot(1,4,1)
sns.histplot(df["sepal_length"],kde=True,color='r')

plt.subplot(1,4,2)
sns.histplot(df["sepal_width"],kde=True,color='g')

plt.subplot(1,4,3)
sns.histplot(df["petal_length"],kde=True,color="y")

plt.subplot(1,4,4)
sns.histplot(df["petal_width"],kde=True,color="b")

plt.show()
```

```python
## plot normalize data

plt.figure(figsize=(15,3))
plt.subplot(1,4,1)
sns.histplot(df2["sepal_length"],kde=True,color='r')

plt.subplot(1,4,2)
sns.histplot(df2["sepal_width"],kde=True,color='g')

plt.subplot(1,4,3)
sns.histplot(df2["petal_length"],kde=True,color="y")

plt.subplot(1,4,4)
sns.histplot(df2["petal_width"],kde=True,color="b")

plt.show()
```
