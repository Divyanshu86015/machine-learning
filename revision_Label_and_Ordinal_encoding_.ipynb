# Label encoding
~ here rank matter bt we don't assign rank

```python
import pandas as pd
```

```python
## create dataframe
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df
```

```python
#perform label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
#create instance of labelencoding
labelencoder=LabelEncoder()
```

```python
labelencoder.fit_transform(df['color'])
```

## Ordinal encoding
~ here rank matter and also assign rank

```python
df=pd.DataFrame({
    "size":['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
## perform Ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
#create an instance of the OrdinalEncoder class and fit_transform
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoded_data= encoder.fit_transform(df[["size"]])
```

```python
# encoded_data conver in dataframe
df1=pd.DataFrame(encoded_data,columns=["size_encoded"])
```

```python
df1
```

```python
# now concat dataframes
pd.concat([df,df1],axis=1)
```

```python
df=pd.DataFrame({
    "feedback":['good','average','poor','average','good','poor'],
    "taste":['good','average','poor','average','good','poor']})
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
#create an instance of the OrdinalEncoder class and fit_transform
encoder=OrdinalEncoder(categories=[['good','average','poor'],['good','average','poor']])
```

```python
encoded_data = encoder.fit_transform(df[['feedback','taste']])
```

```python
encoded_data
```

```python
df1=pd.DataFrame(encoded_data,columns=['feedback_encoded','taste_encoded'])
```

```python
df1
```

```python
pd.concat([df,df1],axis=1)
```
