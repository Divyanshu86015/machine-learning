# Normalization Min Max scaling with sklearn library

```python
import seaborn as sns
```

```python
df=sns.load_dataset("taxis")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

**Now perform  Normalization min max scaling**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit(df[["distance","fare","tip"]])
```

```python
normalized_data= min_max.transform(df[["distance","fare","tip"]])
```

```python
normalized_data
```

```python
## conver normalized data into dataframe
import pandas as pd
```

```python
df1=pd.DataFrame(normalized_data,columns=["distance","fare","tip"])
```

```python
df1
```

```python
#plot data without normalization
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.title("Distance")
sns.histplot(df["distance"],kde=True,color='red')
plt.subplot(1,3,2)
plt.title("Fare")
sns.histplot(df["fare"],kde=True,color='green')
plt.subplot(1,3,3)
plt.title("Tip")
sns.histplot(df["tip"],kde=True,color='blue')
plt.show()
```

```python
## plot normalized data

plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.title("Distance")
sns.histplot(df1["distance"],kde=True,color='red')
plt.subplot(1,3,2)
plt.title("Fare")
sns.histplot(df1["fare"],kde=True,color='green')
plt.subplot(1,3,3)
plt.title("Tip")
sns.histplot(df1["tip"],kde=True,color='blue')
plt.show()
```
