#Covariance And Correlation with Python

```python
import seaborn as sns
```

```python
df =sns.load_dataset("healthexp")
```

```python
df.head()
```

```python
df.tail()
```

```python
## covariance
df.select_dtypes(include=['number']).cov()
```

```python
## person correlation and coefficient
df.select_dtypes(include=['number']).corr(method='pearson')
```

```python
## spearman rank correlation
df.select_dtypes(include=["number"]).corr(method="spearman")
```
